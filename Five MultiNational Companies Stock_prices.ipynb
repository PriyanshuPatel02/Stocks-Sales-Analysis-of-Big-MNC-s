## Microsoft Stock prices prediction

```python

#importing all the required libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.style.use('fivethirtyeight')
```

```python
#importing the dataset
df1 = pd.read_csv("Microsoft.csv")
print(df1.head())
```

```python
#Plotting the graph for better understanding of variations
plt.figure(figsize=(10, 4))
plt.title("Microsoft Stock Prices")
plt.xlabel("Date")
plt.ylabel("Close")
plt.plot(df1["Close"]) #we are ploting graph by mentioning close,high,low,open
plt.show()
```

```python
#Printing the correlations of dataset and visualising it in heatmap
print(df1.corr())
sns.heatmap(df1.corr())
plt.show()
# A correlation heatmap is a graphical representation of a correlation matrix
# representing the correlation between different variables
```

```python
#Setting our x and y variables values for fitting in the model
x1 = df1[["Open", "High", "Low"]]
y1 = df1["Close"]
x1 = x1.to_numpy()
y1 = y1.to_numpy()
y1 = y1.reshape(-1, 1)
```

```python
#dividing the data into training set and test set
from sklearn.model_selection import train_test_split
xtrain1, xtest1, ytrain1, ytest1 = train_test_split(x1, y1, test_size=0.2, random_state=42)
```

```python
#Using Random forest regression model
from sklearn.ensemble import RandomForestRegressor  #supervised ML
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(xtrain1, ytrain1)

# n_estimators : This is the number of trees you want to build before taking
#     the maximum voting or averages of predictions.

# Why random forest Regressor is used?
# A random forest is a meta estimator that fits a number of classifying decision
# trees on various sub-samples of the dataset and uses averaging to
# improve the predictive accuracy and control over-fitting.
```

```python
#predicting the result on test set
ypred1 = regressor.predict(xtest1)
```

```python
#printing the predicted values
data1 = pd.DataFrame(data={"Predicted Rate": ypred1})
print(data1.head())
```
